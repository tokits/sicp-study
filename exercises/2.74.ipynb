### 練習問題2.74
 Insatiable Enterprises, Inc.(貪欲エンタープライズ
社) は、世界中に散らばった多数の独⽴事業所からなる分散型複合
企業である。この会社のコンピュータ設備は、どのユーザーから
⾒てもネットワーク全体がひとつのコンピュータのように⾒える
ような巧妙なネットワークインターフェイス⽅式によって相互接
続されたところだ。貪欲社の社⻑は、初めてネットワークの機能
を使って事業所ファイルから管理情報を取り出そうとして愕然と
した。事務所ファイルはすべて Scheme のデータ構造として実装
されているのに、使われている個々のデータ構造は事業所ごとに
違うのだ。事業所⻑会議が急遽開かれ、事業所の独⽴性をこれま
で通りに保ったままで、本部の要求を満たせるようにファイルを
統合する戦略を探ることになった。
そのような戦略をデータ主導プログラミングによって実装する⽅
法を⽰せ。例として、各事業所の⼈事記録は単独のファイルからな
り、従業員の名前をキーとしたレコードの集合を持っているとす
る。集合の構造は事業所ごとに異なる。さらに、各従業員のレコー
ドはそれ⾃⾝が (事業所ごとに異なる構造を持つ) 集合で、address
と salary のような識別⼦をキーとした情報を含んでいるとする。
具体的には、
a. 指定した⼈事ファイルから指定した従業員のレコードを取得
する get-record ⼿続きを本部向けに実装せよ。この⼿続き
は、任意の事業所のファイルに適⽤できる必要がある。個々
の事業所のファイルはどのように構造化しなければならない
か説明せよ。具体的には、どのような型情報を提供する必要
があるだろうか。
b. 任意の事業所の⼈事ファイル内の与えられた職員のレコード
から給与情報を返す get-salary ⼿続きを本部向けに実装せ
よ。この演算が動くようにするには、レコードはどのように
構造化しなければならないだろうか。
c. 本部向けに、find-employee-record ⼿続きを実装せよ。こ
の⼿続きは、全事業所のファイルから与えられた従業員を検
索し、該当レコードを返す。引数として、従業員名と全事業
所のファイルのリストを取るとせよ。
d. 貪欲社が新しい会社を吸収した場合、新しい⼈事情報を中央
システムに組み⼊れるためにはどのような変更が必要になる
だろうか。

回答は割愛。
