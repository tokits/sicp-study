### 練習問題1.36
fixed-pointを修正し、練習問題 1.22で⽰した newlineとdisplayという基本⼿続きを使って、⽣成する近似値 列を表⽰するようにせよ。

次に、
$x \mapsto \frac{log(1000)}{log(x)}$  
x7→ log(1000)/log(x)の不動点を求めることによって  
$x^x=1000$の解を求めよ(⾃然対数を計算するSchemeの基本⼿続き logを利⽤せよ)。  
平均緩和法を使う場合と使わない場合のステップ数を⽐較せよ。  (fixed-pointを開 始するのに、1という推定値が使えないことに注意。$1$を使うと、 $log(1) = 0$による割り算が発⽣してしまう。) 

```scheme
(define tolerance 0.00001)

(define (fixed-point f first-guess)
  (define (close-enough? v1 v2)
    (begin
     (display v1)
     (display " ")
     (display v2)
     (newline)
     (< (abs (- v1 v2)) tolerance)
     )
    )
  ;tryは予約後になっているようで名前を変えないと動作しなかった
  (define (try-tmp guess)
    ;(display first-guess)
    ;(newline)
    (let ((next (f guess)))
      (if (close-enough? guess next) next
          (try-tmp next)))
    )
  (try-tmp first-guess)
  )
```

```scheme
(import "math")
(fixed-point (lambda (x)(/ (math.log 1000) (math.log x))) 1.1)
```

```scheme
; 平均緩和法
(define (average x y)(/ (+ x y) 2))
(fixed-point
 (lambda (x)(average x (/ (math.log 1000) (math.log x))))
 1.1
 )
```
